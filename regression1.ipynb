```python
# Import the necessary libraries
# Numbers
import numpy as np                                      # linear algebra
import pandas as pd                                     # Data processing, CSV file I/O (e.g. pd.read_csv)

# Plots
import seaborn as sns                                   # importing seaborn for statistical plots
import matplotlib.pyplot as plt                         # matplotlib.pyplot plots data

# Train test split
from sklearn.model_selection import train_test_split   # Sklearn package's randomized data splitting function

# Scaling related
from sklearn import preprocessing                      # Provides several common utility functions: standardization of the data set
from scipy.stats import zscore                         # Computes standard score relative to the sample mean and standard deviation

# Models
from sklearn.linear_model import LinearRegression      # To model the ordinary least squares Linear Regression
from sklearn.linear_model import Ridge                 # To model Linear least squares with l2 regularization
from sklearn.linear_model import Lasso                 # Linear Model trained with L1 prior as regularizer
from sklearn.preprocessing import PolynomialFeatures   # Generates a new feature matrix consisting of all polynomial combinations of the features
from sklearn.linear_model import ElasticNet            # To model Elastic Net Regression

# Metrics
from sklearn import metrics                            # sklearn metrics module implements utility functions to measure classification performance
from sklearn.metrics import mean_squared_error         # Determines mean squared error regression loss
from sklearn.metrics import mean_absolute_error        # Determines mean absolute error regression loss
```

```python
X=df.drop(['strength'],axis=1)
y=df['strength']
```

```python
Xscaled=X.apply(zscore)
Xscaled.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(Xscaled , y ,  test_size=0.30,random_state=1)
print((len(X_train)/len(df.index))*100)
print((len(X_test)/len(df.index))*100)
```

LINEAR REGRESSION

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred_lr=lr.predict(X_test)
print("Linear Regression model co-efficients",(lr.coef_))
print("Linear Regression model intercept ",(lr.intercept_))
print('Mean square error:',mean_squared_error(y_test,y_pred_lr))
print('Root Mean square error:',np.sqrt(mean_squared_error(y_test,y_pred_lr)))
print('mean absolute error:',mean_absolute_error(y_test,y_pred_lr))
```

RIDGE REGRESSION

```python
ridge=Ridge(alpha=.3)
ridge.fit(X_train,y_train)
y_pred_ridge=ridge.predict(X_test)
print("Ridge Regression model co-efficients",(ridge.coef_))
print("Ridge Regression model intercept ",(ridge.intercept_))
print('Mean square error:',mean_squared_error(y_test,y_pred_ridge))
print('Root Mean square error:',np.sqrt(mean_squared_error(y_test,y_pred_ridge)))
print('mean absolute error:',mean_absolute_error(y_test,y_pred_ridge))
```

LASSO REGRESSION

```python
lasso=Lasso(alpha=0.1)
lasso.fit(X_train,y_train)
y_pred_lasso=lasso.predict(X_test)
print("lasso Regression model co-efficients",(lasso.coef_))
print("lasso Regression model intercept ",(lasso.intercept_))
print('Mean square error:',mean_squared_error(y_test,y_pred_lasso))
print('Root Mean square error:',np.sqrt(mean_squared_error(y_test,y_pred_lasso)))
print('mean absolute error:',mean_absolute_error(y_test,y_pred_lasso))
```

ELASTIC REGRESSION

```python
en=ElasticNet(alpha=0.1,l1_ratio=0.5)
en.fit(X_train,y_train)
y_pred_en=en.predict(X_test)
print("elastic Regression model co-efficients",(en.coef_))
print("elastic Regression model intercept ",(en.intercept_))
print('Mean square error:',mean_squared_error(y_test,y_pred_en))
print('Root Mean square error:',np.sqrt(mean_squared_error(y_test,y_pred_en)))
print('mean absolute error:',mean_absolute_error(y_test,y_pred_en))
```
